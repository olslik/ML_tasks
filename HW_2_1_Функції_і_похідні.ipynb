1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Функція залежності оцінки від кількості годин, витрачених на навчання, f(x) = 10*x - 5
def f(x):
  return 10*x - 5

# Похідна функції df(x) = 10
def df(x):
  return [10]*len(x)

xs = np.linspace(0,10,11)

plt.figure(figsize = (10, 6))
plt.plot(xs, f(xs), label = 'f(x) = 10x - 5')
plt.plot(xs, df(xs), label = 'df(x) = 10')

plt.title('Функція залежності оцінки від кількості годин, витрачених на навчання, та її похідна')
plt.legend()
plt.xlabel('Кількість годин, витрачених на навчання')
plt.ylabel('Оцінка')

plt.show()
```

**Висновок.** З кожною додатковою годиною, витраченою на навчання, оцінка збільшується на 10 (про що також додатково свідчить похідна від заданої функції f(x), яка дорівнює константі 10).

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp

# Визначаємо символічну змінну
x = sp.symbols('x')

# Сінусоїда
f_sin = sp.sin(x)
derivative_f_sin = sp.diff(f_sin, x)

# Натуральний логарифм
f_ln = sp.log(x)
derivative_f_ln = sp.diff(f_ln, x)

print("Похідна сінусоїдальної функції sin(x):", derivative_f_sin)
print("Похідна логарифмічної функції ln(x):", derivative_f_ln)
```

```python
x = np.linspace(1,10,100)
f1 = np.sin(x)
df1 = np.cos(x)
f2 = np.log(x)
df2 = 1/x

plt.figure(figsize = (10, 5))
plt.plot(x, f1, color = 'green', linewidth = 3, label = 'f1(x) = sin(x)')
plt.plot(x, df1, '--', color = 'green', label = 'df1(x) = cos(x)')
plt.plot(x, f2, color = 'orange', linewidth = 3, label = 'f2(x) = ln(x)')
plt.plot(x, df2, '--', color = 'orange', label = 'df2(x) = 1/x')

plt.legend()
plt.grid(True)

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
# Визначаємо символічну змінну
x = sp.symbols('x')

# Функція залежності витрат від обсягу виробництва
f_cost = x*x - 5*x + 1
derivative_f_cost = sp.diff(f_cost, x)

print("Похідна функції витрат:", derivative_f_cost)
```

```python
x = np.linspace(0,100,101)
f_cost = x*x - 5*x + 1
df_cost = 2*x - 5

plt.figure(figsize = (10, 5))
plt.plot(x, f_cost, linewidth = 2, label = 'f(x) = x^2 - 5x + 1')
plt.plot(x, df_cost, '--', label = 'df(x) = 2x - 5')

plt.title('Графік залежності функції витрат від кількості вироблених одиниць f(x) та її похідної df(x)')
plt.legend()
plt.show()

```
